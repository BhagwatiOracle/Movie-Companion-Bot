```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies=movies[['movie_id','title','genres','keywords','cast','crew','overview']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies['cast'][0]
```

```python
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id','title','tags','genres']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

**Vectorization - Bag Of Words**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

**Stemming**

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
    return " ".join(y)
```

**Cosine Similarity**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
  movie_index= new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movie_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
from langchain_core.tools import tool
import requests
```

```python
@tool
def get_movie_info(movie:str) -> str:
  """
  This function fetches the movie information from the TMDB API and also recommend movies based on genre.
  """
  movie_id=new_df[new_df['title']==movie]['movie_id'].values[0]
  url=f'https://api.themoviedb.org/3/movie/{movie_id}?language=en-US&api_key=50c43980e4e53178353b20fb2fd086e3'
  response = requests.get(url)
  return response.json()
```

```python
get_movie_info.invoke({'movie':''})
```

```python
@tool
def get_movie_recommendation(movie:str)->list:
  """
  This function recommends movies based on the input movie.
  """
  movie_index= new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  movies=[]
  for i in movie_list:
    movies.append(new_df.iloc[i[0]].title)
  return movies
```

```python
@tool
def get_popular_movies()-> list:
  """
  This function fetches the top 5 popular movies from the TMDB API.
  """
  url='https://api.themoviedb.org/3/movie/popular?language=en-US&page=1&api_key=50c43980e4e53178353b20fb2fd086e3'
  response=requests.get(url)
  results=response.json().get('results')[:5]
  return [movie['title'] for movie in results]
```

```python
get_popular_movies.invoke({'input':'Top popular movies'})
```

```python
@tool
def get_genre_recommendation(genre: str) -> list:
    """
    Recommend top 5 movies from the local dataset based on the input genre.
    Genre should be something like 'Action', 'Comedy', etc.
    """
    genre = genre.strip().lower()

    # Filter movies that contain the genre
    filtered = new_df[new_df['genres'].apply(lambda x: any(genre==genre_item.lower() for genre_item in x))]['title'].head(5).tolist()
    if not filtered:
        return "No movies found for the given genre."

    return filtered
```
